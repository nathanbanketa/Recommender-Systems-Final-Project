```python
import pandas as pd
import numpy as np
import random
from shared import load_data, evaluate_model
random.seed(1)
```

```python
train_df, test_df, restaurants_df = load_data()
all_restaurants = restaurants_df['business_id'].unique().tolist()
test_users = test_df['user_id'].unique().tolist()
```

```python
def generate_random_recs(users, restaurants, k=10):
    recommendations = {}
    for user in users:
        recs = random.sample(restaurants, k)
        recommendations[user] = recs
    return recommendations

random_predictions = generate_random_recs(test_users, all_restaurants, k=10)
```

```python
metrics = evaluate_model(random_predictions, test_df, k=10)

print("Random Baseline Performance:")
print(f"Hit Rate @ 10: {metrics['Hit@10']:.4f}")
print(f"NDCG @ 10:     {metrics['NDCG@10']:.4f}")
```
